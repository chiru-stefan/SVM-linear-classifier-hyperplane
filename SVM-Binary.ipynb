# Import sklearn, numpy and matplotlib

```python
from sklearn import svm
import numpy as np
from matplotlib import pyplot as plt
```

# Generate dummy dataset

```python
X = (np.random.uniform(low=0.0, high=1.0, size=100)*10).reshape(50,2) # reshape to x,y
y = list(map(lambda x: 0 if x[0] <=5 else 1, X)) # create labels
```

# Train Support vector classifier

```python
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

## Separate data in 2 arrays (easier)

```python
plt.clf() # Clear axes

# easier to manage plots with separate arrays
red_dots = X[np.where(X[:, 0] <= 5)] 
blue_dots = X[np.where(X[:, 0] > 5)]
```

# Plot examples and hyperplane

```python
plt.scatter(red_dots[:, 0], red_dots[:, 1], c='red', label='Values where x<=5')
plt.scatter(blue_dots[:, 0], blue_dots[:, 1], c='blue', label='Values where x>5')

# get weights and intercept
w = clf.coef_[0] 
a = -w[0] / w[1]
xx = np.linspace(0, 10) # generate x values
yy = a * xx - (clf.intercept_[0]) / w[1] # y values are calculed based on the angle and intercept

plt.plot(xx, yy, 'k-', c='yellow')

plt.ylim(-1, 11) # limit y values

plt.legend()

plt.tight_layout()
plt.show()
```
